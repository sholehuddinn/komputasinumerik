```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

1. %matplotlib inline: Menampilkan grafik langsung di Jupyter Notebook.
2. import matplotlib.pyplot as plt: Mengimpor Matplotlib untuk membuat grafik.
3. import numpy as np: Mengimpor NumPy untuk operasi numerik, terutama array dan matematika.


```python
rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = 2 * x - 5 + rng.randn(50)
plt.scatter(x, y);
```

1. rng = np.random.RandomState(1):
     - Membuat generator angka acak dengan seed 1 untuk hasil yang dapat direproduksi.
2. x = 10 * rng.rand(50):
     - Membuat 50 nilai acak dalam rentang [ 0 , 10 ) untuk sumbu x.
3. y = 2 * x - 5 + rng.randn(50):
    - Membuat data untuk sumbu y menggunakan persamaan linear 𝑦 = 2 𝑥 − 5, ditambah sedikit noise acak dari distribusi normal.
4. plt.scatter(x, y);:
    - Membuat scatter plot menggunakan pasangan data 𝑥 dan 𝑦.
    - Hasilnya adalah titik-titik yang tersebar sesuai dengan hubungan linear 𝑦 = 2 𝑥 − 5, tetapi dengan variasi akibat noise acak.



```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)

model.fit(x[:, np.newaxis], y)

xfit = np.linspace(0, 10, 1000)
yfit = model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit);
```

1. from sklearn.linear_model import LinearRegression:
    - Mengimpor kelas LinearRegression dari library scikit-learn untuk membuat model regresi linear.
2. model = LinearRegression(fit_intercept=True):
    - Membuat objek model regresi linear dengan opsi untuk menyertakan intercept ( 𝑏 dalam 𝑦 = 𝑚 𝑥+𝑏).
3. model.fit(x[:, np.newaxis], y):
    - Melatih model menggunakan data 𝑥 (diubah menjadi array 2D menggunakan [:, np.newaxis]) dan 𝑦.
4. xfit = np.linspace(0, 10, 1000):
    - Membuat 1000 nilai yang terdistribusi secara merata dalam rentang [0,10] untuk membuat garis regresi yang mulus.
5. yfit = model.predict(xfit[:, np.newaxis]):
    - Menggunakan model yang telah dilatih untuk memprediksi nilai 𝑦 berdasarkan𝑥𝑓 𝑖 𝑡 xfit.
6. Plotting:
    - plt.scatter(x, y): Menampilkan scatter plot data asli.
    - plt.plot(xfit, yfit): Menampilkan garis regresi linear berdasarkan model yang dilatih.



```python
print("Model slope:    ", model.coef_[0])
print("Model intercept:", model.intercept_)
```

1. model.coef_[0]:
    - Nilai kemiringan (slope) dari garis regresi, yaitu 𝑚  dalam persamaan 𝑦 = 𝑚 𝑥+𝑏.
    - Menunjukkan seberapa besar perubahan 𝑦 untuk setiap perubahan 𝑥.
2. model.intercept_:
    - nilai intersep (intercept) dari garis regresi, yaitu 𝑏 dalam persamaan 𝑦 = 𝑚 𝑥+𝑏.
    - Menunjukkan nilai 𝑦 Saat 𝑥 = 0.


```python
rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = np.sin(x) + 0.1 * rng.randn(50)

plt.scatter(x, y)
```

2. x = 10 * rng.rand(50):
  - Membuat 50 angka acak dalam rentang [0,10) untuk sumbu 𝑥.
3. y = np.sin(x) + 0.1 * rng.randn(50):
  - Membuat nilai 𝑦 berdasarkan fungsi sinusoidal 𝑦 = sin( 𝑥 ), ditambah noise acak (0.1 × 0.1× angka acak dari distribusi normal).
4. plt.scatter(x, y):
  - Membuat scatter plot yang menampilkan pasangan data 𝑥 dan 𝑦.
